```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
data = pd.read_csv('./data/IBM_HR_ATTRITION.csv')
```

```python
data['Attrition'] = data.Attrition.replace(["Yes", "No"], [1, 0])
```

```python
sns.lmplot(x='DistanceFromHome',y='Attrition',data = data,hue=None,col=None,row=None)
```

```python
## conclusion : plus on habite loin du travail plus on quitte l'entreprise
```

```python
sns.lmplot(x='YearsWithCurrManager',y='Attrition',data = data,hue=None,col=None,row=None)
```

```python
# plus on reste avec le manager moins on quitte l'entreprise
```
